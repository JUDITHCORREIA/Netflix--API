```python
import re 
my_list=[]
#all_direct=[]
for i in range(len(newdata['titles'])):
    my_dict={}
    types=newdata["titles"][i]["summary"]["type"]
    availability_date=newdata['titles'][i]['jawSummary']['availability']['availabilityDate']
    titles=newdata['titles'][i]['jawSummary']['title']
    release_Years=newdata['titles'][i]['jawSummary']['releaseYear']
    Genre=newdata['titles'][i]['jawSummary']['genres'][0]["name"]
    maturity_description=newdata['titles'][i]['jawSummary']['maturity']['rating']['maturityDescription']
    awards=newdata['titles'][i]['jawSummary']['supplementalMessageIcon']
    
    mg=newdata['titles'][i]['jawSummary']['directors']
    if len(mg)>=1:
        dir_data=mg[0]['name']
        director=dir_data
    else:
        dir_data='unavailable'
        director=dir_data
    

        
    
    
    
    y = (newdata['titles'][i]['jawSummary'])
    e=y.get('runtime')
    if e==None:
        runtimes='unavailable'
    else:
        runtimes=e
    
    x = (newdata['titles'][i]['jawSummary'])
    d=x.get('numSeasonsLabel')
    if d==None:
        seasons="no seasons available"
    else:
        seasons=d
        
    u=(newdata['titles'][i]['jawSummary']['maturity']['rating']['specificRatingReason'])
    if len(u)==0:
        specific_ratings='unavailable'
    else:
         specific_ratings=u
    
    
    
            
            
    a=newdata["titles"][i]["jawSummary"]["cast"]
    castdata=map(lambda f:f['name'],a)
    kk=list(castdata)
    ll=','.join(kk)
    Cast=ll    
            
            
            
            
            
            
            
            
            
            
            
            
            
            
            
    my_dict['Types']=types
    my_dict['Availability_Date']=availability_date
    my_dict['Titles']=titles
    my_dict['Release_years']=release_Years
    my_dict['Genres']=Genre
    my_dict['Maturity_description']=maturity_description
    my_dict['Specific_Rating']=specific_ratings
    my_dict['Awards']=awards
    my_dict['Seasons']=seasons
    my_dict['Runtime']=runtimes
    my_dict['Directors']=director
    my_dict['cast']=Cast
    my_list.append(my_dict)
    
    
```
